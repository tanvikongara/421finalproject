```python
import pandas as pd

# Load the balanced datasets with thresholds of 1000 and 2500
df_smote_1000 = pd.read_csv('../Data/df_smote_1000.csv')
df_smote_2500 = pd.read_csv('../Data/df_smote_2500.csv')
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Creating LabelEncoder for the two datasets and encoding the Label column into a numeric form
le_1000 = LabelEncoder()
df_smote_1000['Label_Encoded'] = le_1000.fit_transform(df_smote_1000['Label'])

le_2500 = LabelEncoder()
df_smote_2500['Label_Encoded'] = le_2500.fit_transform(df_smote_2500['Label'])

# Dropping the original and the encoded label columns to separate for the target variable 
x_1000 = df_smote_1000.drop(['Label', 'Label_Encoded'], axis = 1)
y_1000 = df_smote_1000['Label_Encoded']

x_2500 = df_smote_2500.drop(['Label', 'Label_Encoded'], axis = 1)
y_2500 = df_smote_2500['Label_Encoded']

# Creating a StandardScaler and for normalizing the feature values
standard_scaler_1000 = StandardScaler()
x_1000_scaled = standard_scaler_1000.fit_transform(x_1000)

standard_scaler_2500 = StandardScaler()
x_2500_scaled = standard_scaler_2500.fit_transform(x_2500)

```

```python
from sklearn.model_selection import train_test_split

# Splitting the scaled featured and labels into training and testing sets
x_train_1000, x_test_1000, y_train_1000, y_test_1000 = train_test_split(
    x_1000_scaled, y_1000, test_size = 0.2, random_state = 42, stratify = y_1000)

x_train_2500, x_test_2500, y_train_2500, y_test_2500 = train_test_split(
    x_2500_scaled, y_2500, test_size = 0.2, random_state = 42, stratify = y_2500)
```

```python
from sklearn.decomposition import PCA

# PCA for SMOTE 1000 using 20 components
pca_1000_20 = PCA(n_components = 20, random_state = 42)
x_train_1000_pca_20 = pca_1000_20.fit_transform(x_train_1000)
x_test_1000_pca_20 = pca_1000_20.transform(x_test_1000)

# PCA for SMOTE 2500 using 20 components
pca_2500_20 = PCA(n_components = 20, random_state = 42)
x_train_2500_pca_20 = pca_2500_20.fit_transform(x_train_2500)
x_test_2500_pca_20 = pca_2500_20.transform(x_test_2500)
```

```python
from sklearn.decomposition import PCA

# PCA for SMOTE 1000 using 25 components
pca_1000_25 = PCA(n_components = 30, random_state = 42)
x_train_1000_pca_25 = pca_1000_25.fit_transform(x_train_1000)
x_test_1000_pca_25 = pca_1000_25.transform(x_test_1000)

# PCA for SMOTE 2500 using 25 components
pca_2500_25 = PCA(n_components = 25, random_state = 42)
x_train_2500_pca_25 = pca_2500_25.fit_transform(x_train_2500)
x_test_2500_pca_25 = pca_2500_25.transform(x_test_2500)
```

```python
from sklearn.decomposition import PCA

# PCA for SMOTE 1000 using 30 components
pca_1000_30 = PCA(n_components = 30, random_state = 42)
x_train_1000_pca_30 = pca_1000_30.fit_transform(x_train_1000)
x_test_1000_pca_30 = pca_1000_30.transform(x_test_1000)

# PCA for SMOTE 2500 using 25 components
pca_2500_30 = PCA(n_components = 30, random_state = 42)
x_train_2500_pca_30 = pca_2500_30.fit_transform(x_train_2500)
x_test_2500_pca_30 = pca_2500_30.transform(x_test_2500)
```

```python
# Getting the attack names from the label encoder for easy visualization
attack_names_1000 = le_1000.classes_
attack_names_2500 = le_2500.classes_
```

```python
# Importing the necessary libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# KNN SMOTE 1000

# Initalizing the KNN classifier
knn_1000 = KNeighborsClassifier(n_neighbors = 5)

# Training the model on the dataset and predicting results on the testing set
knn_1000.fit(x_train_1000, y_train_1000)
y_pred_knn_1000 = knn_1000.predict(x_test_1000)

# Creating the classification report for the KNN
print(classification_report(y_test_1000, y_pred_knn_1000, target_names = attack_names_1000))

# Creating the confusion matrix for KNN using sns.heatmap
confusion_matrix_knn_1000 = confusion_matrix(y_test_1000, y_pred_knn_1000)
plt.figure(figsize = (14, 10))
sns.heatmap(confusion_matrix_knn_1000, annot = True, fmt = 'd', cmap = 'Blues',
            xticklabels = attack_names_1000, yticklabels = attack_names_1000)
plt.title('KNN Confusion Matrix (SMOTE 1000)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.tight_layout()
plt.show()

# KNN SMOTE 2500

# Initalizing the KNN classifier
knn_2500 = KNeighborsClassifier(n_neighbors = 5)

# Training the model on the dataset and predicting results on the testing set
knn_2500.fit(x_train_2500, y_train_2500)
y_pred_knn_2500 = knn_2500.predict(x_test_2500)

# Creating the classification report for the KNN
print(classification_report(y_test_2500, y_pred_knn_2500, target_names=attack_names_2500))

# Creating the confusion matrix for KNN using sns.heatmap
confusion_matrix_knn_2500 = confusion_matrix(y_test_2500, y_pred_knn_2500)
plt.figure(figsize = (14, 10))
sns.heatmap(confusion_matrix_knn_2500, annot = True, fmt = 'd', cmap = 'Blues',
            xticklabels = attack_names_2500, yticklabels = attack_names_2500)
plt.title('KNN Confusion Matrix (SMOTE 2500)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.tight_layout()
plt.show()
```
